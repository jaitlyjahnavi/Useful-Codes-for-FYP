```python
country=pd.read_csv('County_2020.csv')
country.head(3)
```

```python
my_map= folium.Map(
    location=[13.133932434766733,16.103938729508073],
    zoom_start=2

)
for _, city in country.iterrows():
 folium.Marker(
    location=[city['Latitude'],city['Longitude']],
    popup=city['Country'],
    tooltip=city['Country']
).add_to(my_map)
my_map
```

```python
import matplotlib.pyplot as ply
import numpy as np
from matplotlib.pyplot import pie, axis, show
```

```python
new_df=country['Country'].value_counts().rename_axis('Country_Count').reset_index(name='Counts')
new_df.head(100)
```

```python
country_name=new_df.Country_Count
values=new_df.Counts
```

```python
fig=ply.figure()
ax=fig.add_axes([1,1,1,1])
ax.axis('equal')


ax.pie(values, labels= country_name,autopct='%1.2f%%')
ply.show()
```
